```python
import random
import datetime
```

```python
# Step 2
def create_sensor(sensor_id, location, sensor_type):
    sensor ={
        "id": sensor_id,
        "location": location,
        "type": sensor_type,
        "value": 0
       
    }
    return sensor
```

```python
sensors = [
    create_sensor("S1", "Living Room", "motion"),
    create_sensor("S2", "Front Door", "door"),
    create_sensor("S3", "Kitchen", "temperature"),
    create_sensor("S4", "Bedroom", "smoke")
]
```

```python
# checkpoint:
print(sensors[0])              
print(sensors[0]["location"])    
print("Number of sensors:", len(sensors))  
```

```python
# Step 3
def check_sensor(sensor, system_mode):
    
    sensor_type = sensor["type"]    
    value = sensor["value"]          

    if sensor_type == "temperature":
        if value < 35:
            return "SAFETY: Temperature too low — frozen pipe risk!"
        elif value > 95:
            return "SAFETY: Temperature too high — equipment failure risk!"

   
    elif sensor_type == "smoke":
        if value == 1:     
            return "SAFETY: Smoke detected — fire risk!"

    
    elif sensor_type == "motion":
        if value == 1 and system_mode == "AWAY":
            return "SECURITY: Motion detected while in AWAY mode!"

    # --- Door rule (security) ---
    elif sensor_type == "door":
        if value == 1 and system_mode == "AWAY":
            return "SECURITY: Front Door opened while in AWAY mode!"

    # If none of the rules matched, everything is fine
    return None
```

```python
# Checkpoint 

#test 1 
test_temp = {"id": "S3", "location": "Kitchen", "type": "temperature", "value": 30}
print(check_sensor(test_temp, "HOME"))


# Test 2 
test_motion = {"id": "S1", "location": "Living Room", "type": "motion", "value": 1}
print(check_sensor(test_motion, "AWAY"))
# Expected: SECURITY: Motion detected while in AWAY mode!

# Test 3 
print(check_sensor(test_motion, "HOME"))
# Expected: None
```

```python
# Step 4

def generate_reading(sensor):
    sensor_type = sensor["type"]

    if sensor_type == "temperature":
        sensor["value"] = random.randint(20, 100)   # a temperature between 20-100°F
    else:
        sensor["value"] = random.randint(0, 1)      # motion/door/smoke → 0 or 1

    return sensor
```

```python
def log_event(message):
    now = datetime.datetime.now().strftime("%H:%M:%S")   
    print(f"[LOG] [{now}] {message}")
```

```python
def trigger_alert(message):
    print(f"[ALERT!] HIGH: {message}")
    log_event("Sending notification to homeowner...")
```

```python
def process_sensor(sensor, system_mode):
    location = sensor["location"]
    sensor_type = sensor["type"]
    value = sensor["value"]

    print(f"[READING] {location} {sensor_type}: {value}")

    alert_message = check_sensor(sensor, system_mode)

    if alert_message != None:
        trigger_alert(alert_message)
```

```python
# Checkpoint
test_sensor = {"id": "S3", "location": "Kitchen", "type": "temperature", "value": 0}

generate_reading(test_sensor)                   
print("Generated value:", test_sensor["value"])  
process_sensor(test_sensor, "AWAY")             
```

```python
# Step 5

class Sensor:

    # Runs automatically when a new sensor is created
    def __init__(self, sensor_id, location, sensor_type):
        self.id = sensor_id
        self.location = location
        self.type = sensor_type
        self.value = 0

    
    def read(self):
        if self.type == "temperature":
            self.value = random.randint(20, 100)
        else:
            self.value = random.randint(0, 1)
        return self.value

   
    def isAbnormal(self, system_mode):
        if self.type == "temperature":
            if self.value < 35:
                return "SAFETY: Temperature too low — frozen pipe risk!"
            elif self.value > 95:
                return "SAFETY: Temperature too high — equipment failure risk!"
            elif system_mode == "HOME" and (self.value < 65 or self.value > 75):
                return "COMFORT: Temperature outside comfort range (65-75°F)"
        elif self.type == "smoke":
            if self.value == 1:
                return "SAFETY: Smoke detected — fire risk!"
        elif self.type == "motion":
            if self.value == 1 and system_mode == "AWAY":
                return "SECURITY: Motion detected while in AWAY mode!"
        elif self.type == "door":
            if self.value == 1 and system_mode == "AWAY":
                return "SECURITY: Front Door opened while in AWAY mode!"
        return None

    
    def reset(self):
        self.value = 0
```

```python
# Checkpoint 
kitchen = Sensor("S3", "Kitchen", "temperature")

print("Location:", kitchen.location)   
print("Starting value:", kitchen.value)

kitchen.read()                         
print("After reading:", kitchen.value)

print("Abnormal?", kitchen.isAbnormal("AWAY"))
```

```python
# Build all sensors as OBJECTS now
sensors = [
    Sensor("S1", "Living Room", "motion"),
    Sensor("S2", "Front Door", "door"),
    Sensor("S3", "Kitchen", "temperature"),
    Sensor("S4", "Bedroom", "smoke")
]

print("Number of sensors:", len(sensors))  
```

```python
# Step 6

for sensor in sensors:
    print(sensor.location)
```

```python
def run_simulation(system_mode, rounds):
    
    
    print("=== HomeGuard Security System ===")
    print("Mode:", system_mode)
    print()

   
    for round_number in range(rounds):
        
    
        now = datetime.datetime.now().strftime("%H:%M:%S")
        print(f"Time: {now}")
        
        for sensor in sensors:
            sensor.read()                       
            print(f"[READING] {sensor.location} {sensor.type}: {sensor.value}")

            alert = sensor.isAbnormal(system_mode)
            if alert != None:
                print(f"[ALERT!]  HIGH: {alert}")
                log_now = datetime.datetime.now().strftime("%H:%M:%S")
                print(f"[LOG] [{log_now}] Sending notification to homeowner...")

        print() 
```

```python
# Step 6

run_simulation("AWAY", 3)
```

```python
# Step 7
# Test 1: Security (AWAY mode)

run_simulation("AWAY", 3)
```

```python
# Test 2: Safety (temperature & smoke)

run_simulation("AWAY", 5) 
```

```python
# Test 3: Comfort (HOME mode)

run_simulation("HOME", 3)
```

Explanation: This program uses a sensor class to represent each home sensor, with methods to generate a random reading and check it against thresholds using if/else logic. The main loop runs several rounds, reads every sensor, and prints a timestamped alert whenever a reading is abnormal for the current mode (HOME or AWAY).
